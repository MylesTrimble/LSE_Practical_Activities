# 3.1.9 Practical activity: Use groupby() and aggregate() functions

## 1. Determine how many films from each year (released from 2012 to the present) were watched on Netflix

```python
# reset_index() will reset the index back to the default 0, 1, 2, etc. indexes.)
# Without the reset_index(), the code snippet will return an error.

mo_gpby = mov_ott.groupby(['Year'])['Netflix'].agg('sum').reset_index()
mo_gpby[mo_gpby['Year'] >= 2012]
```

## 2. What is the average runtime of movies released each year?

```python
mo_gpby1 = mov_ott.groupby(['Year'])['Runtime'].agg('mean').reset_index()
mo_gpby1[mo_gpby1['Year'] >= 2012]
```

## 3. What are the best and worst reviews that movies received on Rotten Tomatoes?

```python
mo_gpby2 = mov_ott.groupby(['Year'])['Rotten Tomatoes'].agg('min', 'max').reset_index()
mo_gpby2[mo_gpby2['Year'] >= 2012]
```

# 3.2.6 Practical activity: Write and apply user-defined functions

The political party has two questions on its mind when it comes to running its ads:

Should the ads be more generic and appeal to as wide an audience as possible?
Should the ads be more provocative and be aired during films with the appropriate age rating?

Mandisa comes up with a few questions she feels might help the political party to make a decision and that she can answer through analysis:

## 1. What is the average rating per movie?

```python
mov_ott_ratings = mov_ott[['ID', 'IMDb', 'Rotten Tomatoes']]

mov_ott_ratings
```

```python
# Replace missing values with 0
mov_ott_ratings_final = mov_ott_ratings.fillna(0)

# View the DataFrame
mov_ott_ratings_final
```

```python
# Write UDF
# Average rating is ((IMDb/10) + Rotten Tomaties)/n
def avg_col(df1, df2):
    df = (df1/10 + df2) / 2
    return df
```

```python
# Add a 'Rating' column to mov_ott_ratings-final df
mov_ott_ratings_final['Rating'] = avg_col(mov_ott_ratings_final['IMDb'],
                                          mov_ott_ratings_final['Rotten Tomatoes'])
```

```python
# View output
mov_ott_ratings_final
```

## 2. How many movies were released per content rating (age)?

```python
# Write UDF
# cat_cnt = category count
def cat_cnt(df1):
    print(df1.value_counts())
```

```python
# Number of movies released per 'Age'
df = mov_ott['Age'].astype('category')

cat_cnt(df)
```

## 3. How many movies were released per year?

```python
# Can apply the same UDF to this question
df_year = mov_ott['Year'].astype('category')

cat_cnt(df_year)
```
